```python
# unzip
import zipfile
with zipfile.ZipFile('/content/drive/MyDrive/zip_file/celeb_data_set_preprocessed_part_0_3.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/current/Dataset')
```

```python
import tempfile
import os
import tensorflow as tf
#maybe change the given path
path = os.path.join(os.getcwd(), "current/Dataset/content/celeb_data_set_preprocessed_part_0_3")
# Save a dataset
print(path)
new_dataset = tf.data.experimental.load(path,compression= 'GZIP')
```

```python
import matplotlib.pyplot as plt
import tensorflow_io as tfio
for img, target in new_dataset.take(2):
  print('a')
  add_dim = tf.zeros((32,128,128,2), dtype=tf.dtypes.float32, name=None)
  img = tf.concat([img,add_dim],3)
  plt.imshow(img.numpy()[0][:][:][:].astype("float32"),cmap='gray', vmin=0, vmax=255)
  plt.show()
  plt.imshow(tfio.experimental.color.lab_to_rgb(target).numpy()[0][:][:][:].astype("float32"))
  plt.show()
```
