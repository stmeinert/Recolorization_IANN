```python
DATADIR = "/content/drive/MyDrive/save_data/27_03/CIC_prob_1024/"

#load dicts

with open(DATADIR+'/metrics_input_target_dict.pickle','rb') as f:
     CIC_input_target_dict = pickle.load(f)

with open(DATADIR+'/metrics_prediction_target_dict.pickle','rb') as f:
     CIC_prediction_target_dict = pickle.load(f)

x_values = np.load(DATADIR+'/x_values.npy')

y_values_CIC_input_target = np.load(DATADIR+'/histogram_data_input_target.npy')
y_values_CIC_prediction_target = np.load(DATADIR+'/histogram_data_prediction_target.npy')

print(x_values)
plt.plot(x_values,y_values_CIC_input_target,'bo',)
print(y_values_CIC_input_target)
print(y_values_CIC_prediction_target)
plt.plot(x_values,y_values_CIC_prediction_target,'go')
plt.show()

#ax = sns.boxplot(x="algo", y="distance", data=CIC_input_target_dict)
```
